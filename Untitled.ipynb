```python
pages = glob("schedulepages/*.pdf")
pages.sort()
```

```python
course = "MINE 603"
lecture = "1L"
df = None
for page in pages:
    currentdf = tabula.read_pdf(page)
    df = pd.concat([df, currentdf])
    print(page)
    newdf = df.loc[df["Course Abbr"] == course]
    if newdf.size == 0: 
        print("Course was not found")
        continue
    print(newdf.size)
    newdf = newdf.loc[newdf["S/T"] == lecture]
    if newdf.size == 0:
        print("Lecture was not found")
        continue
    print(newdf["Cap"])
    break
```
